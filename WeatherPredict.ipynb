# Normalize

```python
from sklearn.preprocessing import StandardScaler

for col in num_columns:
    df[col] = StandardScaler().fit_transform(df[col].values.reshape(-1,1))
```

# train

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    df.drop(columns = ['Weather']).values[:train_end_idx, :],
    df['Weather'].values[:train_end_idx], test_size=0.3)

X_test = df.drop(columns = ['Weather']).values[train_end_idx:, :]
```

```python
# oversampling
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
X_train, y_train = SMOTE(random_state = 42).fit_resample(X_train, y_train)
X_train, y_train = TomekLinks().fit_resample(X_train, y_train)
```

```python
from autosklearn.experimental.askl2 import AutoSklearn2Classifier

automlclassifierV2 = AutoSklearn2Classifier(time_left_for_this_task=2000, per_run_time_limit=80)
automlclassifierV2.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score
autosklearn_y_pred_decision = automlclassifierV2.predict(X_val)
print('Accuracy of Auto-sklearn2: %f' % accuracy_score(y_val, autosklearn_y_pred_decision))
print('f1-score of Auto-sklearn2: %f' % f1_score(y_val, autosklearn_y_pred_decision))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from imblearn.ensemble import BalancedBaggingClassifier

#train tree model
decisionTree_model = DecisionTreeClassifier()
decisionTree_model.fit(X_train,y_train)

svm_model = SVC()
svm_model.fit(X_train,y_train)

lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)

randomForest_model = RandomForestClassifier()
randomForest_model.fit(X_train,y_train)

nb_model = GaussianNB()
nb_model.fit(X_train,y_train)

# max_depth = 1 弱預測器
ada_model = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=200,
    algorithm="SAMME.R", learning_rate=0.5, random_state=42)
ada_model.fit(X_train, y_train)

balBag_model = BalancedBaggingClassifier(base_estimator=DecisionTreeClassifier(),
                                sampling_strategy='not majority',
                                replacement=False,
                                random_state=42)
balBag_model.fit(X_train, y_train)
```

```python
#predict
from sklearn.metrics import accuracy_score, f1_score

decisionTree_y_pred_decision = decisionTree_model.predict(X_val)
print('Accuracy of decision Tree: %f' % accuracy_score(y_val, decisionTree_y_pred_decision))
print('f1-score of decision Tree: %f' % f1_score(y_val, decisionTree_y_pred_decision))

svm_y_pred_decision = svm_model.predict(X_val)  # 0.33001
print('Accuracy of SVM: %f' % accuracy_score(y_val, svm_y_pred_decision))
print('f1-score of SVM: %f' % f1_score(y_val, svm_y_pred_decision))

lr_y_pred_decision = lr_model.predict(X_val)    # 0.370684
print('Accuracy of Logistic Regression: %f' % accuracy_score(y_val, lr_y_pred_decision))
print('f1-score of Logistic Regression: %f' % f1_score(y_val, lr_y_pred_decision))

randomForest_y_pred_decision = randomForest_model.predict(X_val)
print('Accuracy of Random Forest: %f' % accuracy_score(y_val, randomForest_y_pred_decision))
print('f1-score of Random Forest: %f' % f1_score(y_val, randomForest_y_pred_decision))

nb_y_pred_decision = nb_model.predict(X_val)
print('Accuracy of Gaussian NB: %f' % accuracy_score(y_val, nb_y_pred_decision))
print('f1-score of Gaussian NB: %f' % f1_score(y_val, nb_y_pred_decision))

ada_y_pred_decision = ada_model.predict(X_val)
print('Accuracy of AdaBoost: %f' % accuracy_score(y_val, ada_y_pred_decision))
print('f1-score of AdaBoost: %f' % f1_score(y_val, ada_y_pred_decision))

balBag_y_pred_decision = balBag_model.predict(X_val)
print('Accuracy of balance Bagging: %f' % accuracy_score(y_val, balBag_y_pred_decision))
print('f1-score of balance Bagging: %f' % f1_score(y_val, balBag_y_pred_decision))
```
